```python
import numpy as np
import pandas as pd
import dash
import dash_html_components as html #to need to code html components
import dash_core_components as dcc
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc

app = dash.Dash(external_stylesheets=[dbc.themes.DARKLY])
df = pd.read_csv(r'crop2.csv')
df.head(11)
d = pd.read_csv(r'crop1.csv')
d.head(11)

fig1 = go.Figure()

app.layout=html.Div([html.Div(html.H1('Crop Production',style={
                                                            'color':'white', #all should be as string
                                                            'text-align':'center',
                                                            'background':'linear-gradient(135deg,#ffec61,#f321d7)',
                                                            'border': '2px solid white',
                                                            'border-radius': '12px'
                                                            #'display':'inline-block', # display list items horizontally instead
                                                            #'width':'50%' 'border-style': 'solid',
                                                            })),
                    html.Div(html.H2('A Complete Analysis and Visualization of Crop',style={
                                                            'color':'white', #all should be as string
                                                            'text-align':'center',
                                                            'background':'linear-gradient(135deg,#E433FF,#CBFF33)',
                                                            'border-style': 'double'
                                                            })),
                   
                   html.Div(dcc.Dropdown(id='dropdown1',options=[{'label':'Production of Different Agricultural Products','value':'p1'},
                                                             {'label':'Area Wise Production Different Agricultural Products','value':'p2'},
#                                                             {'label':'Costing Bar Chart','value':'p3'},
                                                            {'label':'Cost of Production','value':'p4'},
                                                            {'label':'Cost of Cultivation','value':'p5'},
                                                            {'label':'Doughnut pie','value':'p6'}],
                                                                placeholder="Select an Option",
                                                                style={'width':'100%'})),
                   html.Div([dcc.Graph(id='figure1',figure=fig1,style={'width':'auto',})]),
                   html.Div(html.H2('Presented by: Rudra Pratap Padhi(170301140014) & Subhrajeet Nayak(170301140007)',style={
                                                            'color':'white',
                                                            'font-style': 'oblique',#all should be as string
                                                            'text-align':'center',
                                                            'background':'linear-gradient(135deg,#64b5f6,#f403d1)',
                                                            #'border': '2px solid white',
                                                            'border-radius': '12px',
                                                            'border-style': 'double'                               
                                                            })),
                     
                    ]) 


@app.callback(Output('figure1','figure'),[Input('dropdown1','value')])

def updatefigure(dropdown1):
    global fig1
    if (dropdown1=='p1'):
        
        fig1 = go.Figure([go.Bar(x=df.Crop,y=df.Production_2013_14,name='Production_2013-14'),
                 go.Bar(x=df.Crop,y=df.Production_2014_15,name='Production_2014-15'),
                 go.Bar(x=df.Crop,y=df.Production_2015_16,name='Production_2015-16'),
                 go.Bar(x=df.Crop,y=df.Production_2016_17,name='Production_2016-17'),
                 go.Bar(x=df.Crop,y=df.Production_2017_18,name='Production_2017-18')])
        fig1.update_layout(title="Production of Different Agricultural Products",xaxis_title="crop",yaxis_title="Production in Ton")
        fig1.update_layout(barmode='stack')
        

#         fig1 = go.Figure([go.Bar(x=df.Crop,y=df.Production_2013_14,name='Production_2013-14'),
#                          go.Bar(x=df.Crop,y=df.Production_2014_15,name='Production_2014-15'),
#                          go.Bar(x=df.Crop,y=df.Production_2015_16,name='Production_2015-16'),
#                          go.Bar(x=df.Crop,y=df.Production_2016_17,name='Production_2016-17'),
#                          go.Bar(x=df.Crop,y=df.Production_2017_18,name='Production_2017-18')])
#         fig1.update_layout(title="Production of Different Agricultural Products",xaxis_title="crop",yaxis_title="Production in quintal")
#         fig1.update_layout(barmode='group')

    if (dropdown1=='p2'):
        fig1 = go.Figure([go.Bar(x=df.Crop,y=df.Area_2013_14,name='Area_2013_14'),
                         go.Bar(x=df.Crop,y=df.Area_2014_15,name='Area_2014-15'),
                         go.Bar(x=df.Crop,y=df.Area_2015_16,name='Area_2015-16'),
                         go.Bar(x=df.Crop,y=df.Area_2016_17,name='Area_2016-17'),
                         go.Bar(x=df.Crop,y=df.Area_2017_18,name='Area_2017-18')])
        fig1.update_layout(title="Area Wise Production Different Agricultural Products",xaxis_title="crop",yaxis_title="hectare")
        fig1.update_layout(barmode='group')
        
#     if (dropdown1=='p3'):
#         fig1 = go.Figure([go.Bar(x=d.State,y=d.Cost_of_Cultivation,name='Cost of Cultivation'),
#         go.Bar(x=d.State,y=d.Cost_of_Production,name='Cost of production')])
#         fig1.update_layout(title="Group Bar Chart",xaxis_title="State",yaxis_title="cost")
#         fig1.update_layout(barmode='group')
#         fig1 = go.Figure([go.Bar(x=d.State,y=d.Cost_of_Cultivation,name='Cost of Cultivation'),
#                 go.Bar(x=d.State,y=d.Cost_of_Production,name='Cost of production')])
#         fig1.update_layout(title="Costing Bar Chart",xaxis_title="State",yaxis_title="cost")
#         fig1.update_layout(barmode='group')
        
    if (dropdown1=='p4'):
        fig1 = go.Figure(data=[go.Pie(labels=d.State, values=d.Cost_of_Production,textinfo='label+percent',insidetextorientation='radial')])
        fig1.update_layout(title_text="Cost of Production ")
    
        
    if (dropdown1=='p5'):
        fig1 = go.Figure(data=[go.Pie(labels=d.State, values=d.Cost_of_Cultivation,textinfo='label+percent',insidetextorientation='radial')])
        fig1.update_layout(title_text="Cost of Cultivation")
        


    if (dropdown1=='p6'):

        fig1 = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
        fig1.add_trace(go.Pie(labels=d.State, values=d.Cost_of_Cultivation, name="Cost of cultivation"),1, 1)
        fig1.add_trace(go.Pie(labels=d.State, values=d.Cost_of_Production, name="Cost of Production"),1, 2)
        fig1.update_traces(hole=.3, hoverinfo="label+percent+name")
        fig1.update_layout( title_text="Information",
        annotations=[dict(text='cultivation', x=0.2, y=0.5, font_size=15, showarrow=False),
        dict(text='Production', x=0.82, y=0.5, font_size=15, showarrow=False)])
        
    return fig1
        


if __name__ == '__main__':         
    app.run_server()
```
